```python
# select records from the patient_encouonter table
connection_string = f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}"
try:
    engine = create_engine(connection_string)

    with engine.connect() as conn:
        patient_encounter = pd.read_sql('''SELECT patient_nbr,encounter_id
                                            ,admission_type_id
                                            ,admission_source_id
                                            ,discharge_disposition_id
                                            ,time_in_hospital,payer_code,medical_specialty,num_lab_procedures,num_procedures
                                            ,num_medications,number_outpatient,number_emergency,number_inpatient
                                            ,pe.diag_1,icd9_1.icd9_description AS diag_1_desc
                                            ,pe.diag_2,icd9_2.icd9_description AS diag_2_desc
                                            ,pe.diag_3,icd9_3.icd9_description AS diag_3_desc
                                            ,number_diagnoses
                                            ,max_glu_serum,"a1cresult",metformin,repaglinide,nateglinide,chlorpropamide,acetohexamide
                                            ,glimepiride,glipizide,glyburide,tolbutamide,tolazamide,examide,pioglitazone,rosiglitazone
                                            ,troglitazone,acarbose,miglitol,citoglipton,insulin,"glyburidemetformin","glipizidemetformin"
                                            ,"glimepiridepioglitazone","metforminrosiglitazone","metforminpioglitazone"
                                            ,change_in_meds,"diabetesmed",readmitted
                                            FROM patient_encounter pe
                                            INNER JOIN icd9 icd9_1
                                                ON TO_NUMBER(pe.diag_1, 'L999.99') BETWEEN icd9_1.code_start_int AND icd9_1.code_end_int
                                                    OR pe.diag_1 BETWEEN icd9_1.code_start_str AND icd9_1.code_end_str
                                            INNER JOIN icd9 AS icd9_2
                                                ON TO_NUMBER(pe.diag_2, 'L999.99') BETWEEN icd9_2.code_start_int AND icd9_2.code_end_int
                                                    OR pe.diag_2 BETWEEN icd9_2.code_start_str AND icd9_2.code_end_str
                                            INNER JOIN icd9 AS icd9_3
                                                ON TO_NUMBER(pe.diag_3, 'L999.99') BETWEEN icd9_3.code_start_int AND icd9_3.code_end_int
                                                    OR pe.diag_3 BETWEEN icd9_3.code_start_str AND icd9_3.code_end_str
                                        ;''', con = conn)

        conn.close()

except (Exception) as error:
        error_log   = 'Error select data from patient_encounter table. Check system file.'                  
        with open('error_log.txt','w') as error:
           error.write(error_log)


patient_encounter.head()
```

```python
# merge admission_source and patient_encounter
patient_encounter_df = pd.merge(patient_encounter, admission_source_df, on='admission_source_id', how='inner')
```

```python
# merge admission_type and patient_encounter
patient_encounter_df = pd.merge(patient_encounter_df, admission_type_df, on='admission_type_id', how='inner')
```

```python
# merge discharge_disposition_df and patient_encounter
patient_encounter_df = pd.merge(patient_encounter_df, discharge_disposition_df, on='discharge_disposition_id', how='inner')
```

```python
# merge patient and patient_encounter
patient_all_df = pd.merge(patient, patient_encounter_df, on='patient_nbr', how='inner')
```

```python
# perform analysis

# patient per admission types (Why) : represents the priority and nature of the patient's arrival, often used for scheduling and administrative analysis.
patient_admission_type = patient_all_df.groupby('admission_type_description')['patient_nbr'].count().sort_values(ascending=False).reset_index()
patient_admission_type
```

```python
# top 3 admission types
df_3 = patient_admission_type.head(3)
fig = px.bar(df_3, x = 'admission_type_description'
             ,y='patient_nbr', title='Admission Types Distribution'
             , color='admission_type_description'
             , labels={'patient_nbr': 'patients', 'admission_type_description': 'descriptions'})
fig.update_layout(showlegend=False)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```

```python
# patient per admission source : (Where) - identifies the location or context of the patient immediately prior to being admitted to the facility
patient_admission_source = patient_all_df.groupby('admission_source_description').agg({
    'patient_nbr': 'count',
    'time_in_hospital': 'mean',
    }).sort_values('patient_nbr', ascending=False)
patient_admission_source = patient_admission_source.reset_index()
patient_admission_source
```

```python
# visualize top 5
fig = px.bar(patient_admission_source.head(), x='admission_source_description'
             , y='patient_nbr', color='time_in_hospital'
             , barmode='group'
             , title='Admission Source Distribution')

fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)

fig.show()
```

```python
# visualize the top 5
df_5 = patient_admission_source.head()
fig = px.pie(df_5, values='patient_nbr', names='admission_source_description'
             ,  title='Admission Sources Distribution')
fig.show()
```

```python
# Top Ten Primary Diagnosis Codes
connection_string = f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}"
try:
    engine = create_engine(connection_string)

    with engine.connect() as conn:
        primary_top_ten_df = pd.read_sql('''WITH patient_primary_diagnosis_cte AS (
                                            SELECT pe.patient_nbr
                                                ,diag_1
                                                ,icd9_1.icd9_description
                                            FROM patient_encounter pe
                                            INNER JOIN icd9 icd9_1
                                                ON TO_NUMBER(pe.diag_1, 'L999.99') BETWEEN icd9_1.code_start_int AND 
                                                        icd9_1.code_end_int
                                                    OR pe.diag_1 BETWEEN icd9_1.code_start_str AND 
                                                        icd9_1.code_end_str
                                            )
                                            SELECT diag_1 AS primary_diagnosis
                                                , icd9_description AS description
                                                , COUNT(ppdc.patient_nbr) AS num_of_patients
                                            FROM patient_primary_diagnosis_cte ppdc
                                            GROUP BY diag_1
                                                ,icd9_description
                                            ORDER BY COUNT(ppdc.patient_nbr) DESC 
                                            LIMIT 10;
                                        ''', con = conn)

        conn.close()

except (Exception) as error:
        error_log   = 'Error select data from patient_encounter table. Check system file.'                  
        with open('error_log.txt','w') as error:
           error.write(error_log)


primary_top_ten_df.head(10)
```

```python
# visualize
fig = px.bar(primary_top_ten_df, x='primary_diagnosis',
             y='num_of_patients', color='num_of_patients'
             , title='Top Ten Primary Diagnosis Codes'
             , hover_data='description')

fig.show()
```

```python
# patient per discharge disposition types : the final destination or setting a patient goes to upon leaving a healthcare facility,
# determining their next level of care.  It defines whether the patient is going home (with or without support), 
# transferring to another facility (like nursing or rehab), or leaving against medical advice
patient_discharge_disposition_type = patient_all_df.groupby('discharge_disposition_description').agg({
    'patient_nbr': 'count'
}).sort_values(by='patient_nbr', ascending=False)
patient_discharge_disposition_type = patient_discharge_disposition_type.reset_index()
patient_discharge_disposition_type['total_patients'] = patient_discharge_disposition_type['patient_nbr'].sum()
patient_discharge_disposition_type['pct_discharge_disposition'] = ((patient_discharge_disposition_type['patient_nbr'] / patient_discharge_disposition_type['patient_nbr'].sum()) * 100).round(1)
patient_discharge_disposition_type
```

```python
# visualize discharge_disposition_description : Top Three
fig = px.bar(patient_discharge_disposition_type.head(3), x='discharge_disposition_description'
             ,y='patient_nbr', hover_data='pct_discharge_disposition', color='discharge_disposition_description'
            ,title='Discharge Disposition Distribution')

fig.update_layout(showlegend=False)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)

fig.show()
```

```python
# Determine the number of readmitted patients and the percentage of total encounters that they represent
readmitted_df = patient_all_df.groupby('readmitted').agg({
    'encounter_id': 'count',
    'patient_nbr': 'nunique'
    }) #.sort_values(by='encounter_id', ascending=False)
readmitted_df = readmitted_df.reset_index()
readmitted_df['total_encounters'] = readmitted_df['encounter_id'].sum()
readmitted_df['percent_encounters'] = ((readmitted_df['encounter_id'] / readmitted_df['total_encounters']) * 100).round(1)
readmitted_df
```

```python
# visualize
fig = px.bar(readmitted_df, x='readmitted', y='encounter_id', color='encounter_id'
             , title='Readmittance Encounters'
             , hover_data='percent_encounters')
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```
